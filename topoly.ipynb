```python
from topoly import *
```

```python
proteins = []
with open("PROT_DATA/PF04034_reduced.fasta", "r") as f:
    lines = f.readlines()
    for line in lines:
        if line.strip("\n").startswith(">"):
            numba = line.find("(")
            numbb = line.find(")")
            proteins.append(line[numba+1:numbb])
len(proteins)
```

```python
score = []
for i in range(len(proteins)):
    with open(f"PROT_DATA/cif/AF-{proteins[i]}-F1-model_v3.cif", "r") as x:
        lines = x.readlines()
        for line in lines:
            if line.startswith("_ma_qa_metric_global.metric_value"):
                score.append(line[-6:-1])
len(score)
```

```python
lista = []
seq = ''
with open("PROT_DATA/PF04034_reduced.fasta", "r") as file:
    for line in file:
        if '>' in line:
            lista.append((id, seq))
            id =  line[line.index('(')+1:line.index(')')]
            seq = ''
        else:
            seq = seq + line
```

```python
h = []
for i in range(len(proteins)):
  h.append(homfly(f'PROT_DATA/cif/AF-{proteins[i]}-F1-model_v3.cif', max_cross=20, tries=300, translate=True))
print(h[0])
```

```python
a = 0
for i in range(len(proteins)):
  # print(max(h[i]))
  a += 1
print(a)
```

```python
with open("PF04034_details.txt", "w") as file:
    for i in range(1,len(lista)):
        file.write(str(lista[i][0]) + "," + str(lista[i][1]).replace("\n", "") + "," + score[i-1] + "," + str(max(h[i-1])) + "\n")
```
